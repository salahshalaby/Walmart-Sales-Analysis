```python
df['Month'] = df['Date'].dt.month
```

```python
monthly_sales = (
    df.groupby('Month')['Weekly_Sales']
      .sum()
      .sort_values(ascending=False)
)

monthly_sales
```

```python
monthly_avg = (
    df.groupby('Month')['Weekly_Sales']
      .mean()
      .sort_values(ascending=False)
)

monthly_avg
```
